```python
def step_gradient(data,m,c,a):
    l=len(data)#len(data) gives no of rows in the array
    slope_wrt_m=0
    slope_wrt_c=0
    for i in range(l):
        x=data[i,0]
        y=data[i,10]
        slope_wrt_m+=(-2/l)*(y-m*x-c)*x
        slope_wrt_c+=(-2/l)*(data[i,10]-m*data[i,0]-c)
    m=m-(a*slope_wrt_m)
    c=c-(a*slope_wrt_c)
    return m,c
        
```

```python
def cost(data,m,c):
    l=len(data)
    cost=0
    for i in range(l):
        x=data[i,0]
        y=data[i,10]
        cost+=(1/l)*((y-m*x-c)**2)
    return cost
```

```python
def gradient_descent(data,m,c,a,num_ite):
    for i in range(num_ite):
        print(cost(data,m,c))
        m,c=step_gradient(data,m,c,a)
    return m,c
    
```
