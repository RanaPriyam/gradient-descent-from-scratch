```python
def step_gradient(data,m,a):
    l=len(data)#len(data) gives no of rows in the array
    slope=np.zeros(11)
    for j in range(11):
        for i in range(l):
            x=data[i,:11]
            y=data[i,11]
            slope[j]+=(-2/l)*(y-(m*x).sum())*x[j]
        m[j]-=a*slope[j]
    return m
        
```

```python
def cost(data,m):
    l=len(data)
    cost=0
    for i in range(l):
        x=data[i,:11]
        y=data[i,11]
        cost+=(1/l)*((y-(m*x).sum())**2)
    return cost
```

```python
def gradient_descent(data,m,a,num_ite):
    for i in range(num_ite):
        print(cost(data,m))
        m=step_gradient(data,m,a)
    return m
    
```
